Knowing the degree to which the data properties are related to each other, we can use the corr function of the data frame to get a table showing the degree to which the numeric properties are related to each other:

```python
df.corr()
```

The degree of correlation ranges from 0 to 1, where the closer the degree is to the value 1, the greater the correlation between the two properties, and this allows us to dispense with one of them, and this is useful in the topic of dimensionality reduction, and the closer the degree to the value 0, the more the correlation between the two properties decreases and becomes They are independent of each other so that we cannot do without one of them.

To make the previous table visible, we do the following:

```python
corr = df.corr()   
fig, ax = plt.subplots(figsize=(10, 10))
ax.matshow(corr)  
plt.xticks(range(len(corr.columns)), corr.columns)  
plt.yticks(range(len(corr.columns)), corr.columns)
```

In the previous figure, the closer the color to black, the higher the correlation, and the closer the color to white, the lower the correlation.

We notice from the figure that there is a relationship between the number of times the content of the material is visited and the number of times the hand is raised in the class, where the more one of them increases the other and vice versa.
in another words,Observing all data points, we can see that there is a strong relationship between rasinghands and vistedresouses. In general, the higher the visiting resourses of a student, the higher they are rasing thier hands.
