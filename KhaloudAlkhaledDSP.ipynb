8-In many experiments, we need to obtain a graphical figure showing the number of recurrences of records for a particular property (nominal or ordinal) according to the range of values ​​it contains. The Seaborn library provides a countplot function that does this. For example, if we want to know the numbers of students by subject, we do the following.

```python
fig = sns.countplot(x="Topic", data=df, palette="muted");
fig.set_xticklabels( fig.get_xticklabels() ,  rotation=70)
plt.show()
```

9-Displaying the numbers of students by subjects with the breakdown of numbers by degree We pass the hue parameter countplot function and specify the name of the degree property as follows

```python
plt.subplots(figsize=(13,10))
fig = sns.countplot(x="Topic", data=df, palette="Set2",hue="Class");
fig.set_xticklabels( fig.get_xticklabels() ,  rotation=70)
plt.show()
```

13-One of the fastest ways to help us understand data and get quick readings and notes about it is to use data visualization techniques, and the various figures, graphs and statistics that can be made. We will mention some of them under this heading.

One of the important topics during data analysis and exploration is to know the degree of correlation of data properties with each other, and we can use the corr function of the data frame to obtain a table showing the degree of correlation of numerical properties to each other:

```python
df.corr()
```

14-The degree of correlation ranges from 0 to 1, where the closer the degree is to the value 1, the greater the correlation between the two properties, and this allows us to dispense with one of them, and this is useful in the topic of dimensionality reduction, and the closer the degree to the value 0, the more the correlation between the two properties decreases and becomes They are independent of each other so that we cannot do without one of them.

To make the previous table visible, we do the following:

```python
corr = df.corr()   
fig, ax = plt.subplots(figsize=(10, 10))
ax.matshow(corr)  
plt.xticks(range(len(corr.columns)), corr.columns)  
plt.yticks(range(len(corr.columns)), corr.columns)
```

In the previous figure, the closer the color to black, the higher the correlation, and the closer the color to white, the lower the correlation.
